```python
# 실시간 상황 정보 가져오는 코드
area = AREA.index.tolist()
area_df = pd.DataFrame()

for a in area:
    AREA_NM = a
    url = f'http://openapi.seoul.go.kr:8088/52775445616b737939387643596a70/json/citydata_ppltn/1/10/{AREA_NM}'
    json_data  = requests.get(url).json()
    tmp = pd.DataFrame(json_data.get('SeoulRtd.citydata_ppltn'))
    tmp.set_index('AREA_NM', inplace=True)
    area_df = pd.concat([area_df, tmp])
```
